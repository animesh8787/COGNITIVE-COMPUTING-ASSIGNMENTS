```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
#     Generate x values from -10 to 10 using np.linspace().

import numpy as np
x=np.linspace(-10,10)
print(x,'\n')

#     Compute y values for:
#       y = M⋅x^2
#       y = M⋅sin(x)

M=int(input("Enter M:"))
y1=M*x**2
print('Y1=M*x^2\n',y1,'\n')
y2=M*np.sin(x)
print('Y2=M*sin(x)\n',y2,'\n')

# Plot both functions in a single figure:
#   Use different colors and line styles.
#   Add legend, grid, and title.


plt.plot(x,y1,'r--',label='Y1=M*x^2')
plt.plot(x,y2,'b-',label='Y2=M*sin(x)')
plt.legend()
plt.grid()
plt.title('Mathematical Functions')
plt.show()
```

```python
# Q.2 Create a dataset of five subjects and scores.
import pandas as pd
data={
    'Subject':['cognitive','prob and stat','dbms','ai','daa'],
    'Score':[87,88,96,86,89]
}
print(data)
#     Convert it into a Pandas DataFrame.
df=pd.DataFrame(data)
print(df)

#     Plot the scores using a Seaborn bar plot with:
sns.set_style('darkgrid')

#       Different colors for each bar.
ax=sns.barplot(x='Subject',y='Score',data=df,palette='deep',hue='Subject',legend=False)

#       Annotations on top of each bar.
for index, row in df.iterrows():
    ax.text(index, row.Score + 1, str(row.Score), ha='center')

#       Title, axis labels, and grid.
plt.title('Subjects vs Scores')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.grid(axis='y')
plt.show()
```

```python
# Q3. Write a Python script to select your roll number as seed for NumPy and
# then generate a dataset of 50 values using np.random.randn().
roll_number=102317101
np.random.seed(roll_number)

data=np.random.randn(50)
print(data)
# Create a 2x2 subplot layout displaying:
fig,axes = plt.subplots(2,2,figsize=(10, 8))

#   Line plot showing cumulative sum.
axes[0,0].plot(np.cumsum(data))
axes[0,0].set_title('Line plot showing cumulative sum.')
axes[0,0].set_xlabel('Indexes')
axes[0,0].set_ylabel('Cumulative Sums')
axes[0,0].grid(True)
#   Scatter plot with random noise.
x=np.linspace(0,10,50)
print(x)

axes[0,1].scatter(x,data)
axes[0,1].set_title('Scatter plot with random noise.')
axes[0,1].set_xlabel('X values')
axes[0,1].set_ylabel('Noise')
axes[0,1].grid(True)

# Customize with titles, axis labels, and grids.
axes[1,0].axis("off")
axes[1,1].axis("off")
```
